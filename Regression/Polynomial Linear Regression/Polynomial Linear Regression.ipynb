```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df =pd.read_csv('Position_Salaries.csv')
df.head()
```

```python
X = df.iloc[:, 1:-1].values
print(X)
y = df.iloc[:, -1].values
print(y)
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
poly_regressor = PolynomialFeatures(degree=2)
X_poly = poly_regressor.fit_transform(X)
lr2 = LinearRegression()
lr2.fit(X_poly, y)
```

```python
linear_reg_pred = lr.predict(X)

plt.scatter(X, y, color='red')
plt.plot(X, linear_reg_pred, color='blue')
plt.title('Linear Regression')
```

```python
poly_linear_reg_pred = lr2.predict(X_poly)

plt.scatter(X, y, color='red')
plt.plot(X, poly_linear_reg_pred, color='green')
plt.title('Polynomial Linear Regression')
```

```python
new_poly_regressor = PolynomialFeatures(degree=4)
X_poly = new_poly_regressor.fit_transform(X)

lr3 = LinearRegression()
lr3.fit(X_poly, y)

new_poly_linear_reg_pred = lr3.predict(X_poly)

plt.scatter(X, y, color='red')
plt.plot(X, new_poly_linear_reg_pred, color='purple')
plt.title('New Polynomial Linear Regression')
```

```python
lr.predict([[6.5]])
```

```python
ln2 = lr2.predict(poly_regressor.fit_transform([[6.5]]))
print(ln2)
ln3 = lr3.predict(new_poly_regressor.fit_transform([[6.5]]))
print(ln3)
```

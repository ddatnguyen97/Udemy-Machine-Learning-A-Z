```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df =pd.read_csv('./Dataset/Ads_CTR_Optimisation.csv')
```

```python
N = 10000
d = 10
ads_selected = []
num_of_reward_1 = [0] * d
num_of_reward_0 = [0] * d
total_reward = 0

for n in range(0, N):
    ad = 0
    max_random_beta = 0
    for i in range(0, d):
        random_beta = random.betavariate(num_of_reward_1[ad] + 1, num_of_reward_0[ad] + 1)
        if random_beta > max_random_beta:
            max_random_beta = random_beta
            ad = i
    ads_selected.append(ad)
    rewards = df.values[n, ad]
    if rewards == 1:
        num_of_reward_1[ad] += 1
    else:
        num_of_reward_0[ad] += 1
    total_reward += rewards
```

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
```
